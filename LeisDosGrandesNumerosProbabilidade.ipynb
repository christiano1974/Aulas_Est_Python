```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random as rd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 100 lançametos de uma moeda honesta 0 para cara e 1 coroa
resultados = []
for k in range(200):
    resultados.append(rd.randint(0,1))
```

```python
# resultados dos 200 lançamentos.
print(resultados)
```

```python
# frequências relativas em relação a cara.
f_cara = []
for k in range(200):
    soma = 0
    for n in range(k):
        if resultados[n] == 0:
            soma += 1
    f_cara.append(np.round(soma/(k + 1),2))
print(f_cara)
```

```python
# Gráfico das leis dos grandes números aplicados na modelagem da probabilidade do número de caras
# no lançamento de uma moeda.
plt.figure(figsize=(15,4))
plt.plot(f_cara,color = 'red',marker = '*');
plt.title('Frequência relativa (Probabilidade de número de caras)')
plt.xlabel('Número de repetições (lançamentos)');
plt.ylabel('Frequência Relativa');
plt.grid(True)
plt.show()
```
