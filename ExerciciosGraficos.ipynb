```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# exercício número 1
data = {'nº de drogas':np.arange(7),'Qtde de estudantes':[140,100,400,80,180,50,50],'freq. Relativa Acumulada':[0.14,0.24,0.64,0.72,0.90,0.95,1]}
df = pd.DataFrame(data)
df
```

```python
# cálcula da média,mediana e moda para dados agrupados
soma = 0
for linha in range(7):
    soma += df.iloc[linha,0]*df.iloc[linha,1]
media = soma/sum(df['Qtde de estudantes'])
media
mediana = 2
moda = 2
print('O valor da média do uso de drogas é igual a {:5.2f}'.format(media))
print('O valor da mediana do uso de drogas é igual a {:5.2f}'.format(mediana))
print('O valor da moda do uso de drogas é igal a {:5.2f}'.format(moda))
```

```python
plt.figure(figsize = (10,5))
df['Qtde de estudantes'].plot(kind = 'bar',rot = 0)
plt.xlabel('Qtde de drogas consumidas')
plt.ylabel('Freq. Absoluta')
plt.title('Consumo de drogas no campus.')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.pie(df['Qtde de estudantes'],autopct = '%1.1f%%')
plt.legend(labels = df['nº de drogas'],loc='center right',bbox_to_anchor = (1.2,0.5))
plt.title("Porcentagens de nº de drogas usadas pelos estudantes do campos")
plt.show()
```

```python
# exercício número 3
dados = np.array([70,79,78,65,76,80,63,76,74,72,69,72,81,75,77,83,74,81,66,85,82,75,86,80,80,71,80,75,64,80,87])
len(dados)
```

```python
amplitudeTotal = dados.max() - dados.min()
numeroClasses = np.sqrt(len(dados))
amplitudeClasse = round(amplitudeTotal/numeroClasses,1)
print('R = {:.0f}'.format(amplitudeTotal))
print('k = {:.0f}'.format(numeroClasses))
print(f'h = {amplitudeClasse}')
```

```python
print('valor min = {:.0f} e valor max = {:.0f}'.format(dados.min(),dados.max()))
print('vou considerar 5 classes')
amplitudeClasse = round(amplitudeTotal/5,1)
print(f'nova amplitude de classe {amplitudeClasse}')
```

```python
classes = ['[ 63.0 ; 67.8 [','[ 67.8 ; 72.6 [','[ 72.6 ; 77.4[','[ 77.4 ; 82.2 [','[ 82.2 ; 87.0 ]']
freqAbsolutas = []
freqAbsolutas.append(len(dados[dados < 67.8]))
freqAbsolutas.append(len(dados[dados < 72.6]) - len(dados[dados< 67.8]))
freqAbsolutas.append(len(dados[dados < 77.4]) - len(dados[dados< 72.6]))
freqAbsolutas.append(len(dados[dados < 82.2]) - len(dados[dados< 77.4]))
freqAbsolutas.append(len(dados[dados <= 87.0]) - len(dados[dados< 82.2]))
freqAbsolutas
```

```python
tabela = pd.DataFrame({'classes':classes,'freq. Absoluta':freqAbsolutas},index = [1,2,3,4,5])
tabela['freq. Acumulada'] = np.cumsum(freqAbsolutas)
tabela['freq Relativa (%)'] = [round(freq/31*100,1) for freq in freqAbsolutas]
tabela
```

```python
plt.figure(figsize = (10,5))
ax = plt.hist(dados,bins = [63.0,67.8,72.6,77.4,82.2,87.0],edgecolor = 'black',stacked = True)
plt.title('histograma da dieta de engorda dos frangos')
plt.ylabel('Freq. Absoluta')
plt.xlabel('Peso dos frangos Kg')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.boxplot(dados)
plt.ylabel('Peso dos frangos')
plt.title('Box-plot da dieta de engorda dos frangos')
plt.show()
```

```python
# exercício 4
classes = ['[0,20[','[20,40[','[40,80[','[80,100[','[100,120[','[120,140[','[140,180[','180 < ']
freqAbsFumantes = [78,133,142,206,197,220,151,412]
freqRelFumantes = [round(valor/sum(freqAbsFumantes),3)*100 for valor in freqAbsFumantes]
freqAbsNaoFumantes = [3300,72,23,15,7,8,9,11]
freqRelNaoFumantes = [round(valor/sum(freqAbsNaoFumantes),3)*100 for valor in freqAbsNaoFumantes]
tabelaFreqFumantes = pd.DataFrame({'classes':classes,'freq. Absoluta':freqAbsFumantes,'freq. Relativa (%)':freqRelFumantes})
tabelaFreqNaoFumantes = pd.DataFrame({'classes':classes,'freq. Absoluta':freqAbsNaoFumantes,'freq. Relativa (%)':freqRelNaoFumantes})
print('tabela de fumantes')
print(tabelaFreqFumantes)
print('tabela não fumantes')
print(tabelaFreqNaoFumantes)
```

```python
# histogramas para os grupos (fumante e não fumante)
plt.figure(figsize = (15,12))
plt.subplot(2,1,1)
plt.bar(tabelaFreqFumantes['classes'],tabelaFreqFumantes['freq. Relativa (%)'],color = 'red',width = 1,edgecolor = 'black')
plt.plot(range(8),tabelaFreqFumantes['freq. Relativa (%)'],color = 'black',linestyle = '--')
plt.title("Histograma dos níveis de nicotinina para fumantes")
plt.xlabel("Nível de nicotinina")
plt.ylabel("Freq. Relativa")
plt.legend(labels = ['polígono de frequência','histograma'],loc = 'upper left')
plt.subplot(2,1,2)
plt.bar(tabelaFreqFumantes['classes'],tabelaFreqNaoFumantes['freq. Relativa (%)'],color = 'blue',width = 1,edgecolor = 'black')
plt.plot(range(8),tabelaFreqNaoFumantes['freq. Relativa (%)'],color = 'black',linestyle = '--')
plt.title("Histograma dos níveis de nicotinina para não fumantes")
plt.xlabel("Nível de nicotinina")
plt.legend(labels = ['polígono de frequência','histograma'],loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize = (12,10))
plt.subplot(2,1,1)
plt.plot(range(8),tabelaFreqFumantes['freq. Relativa (%)'],color = 'red')
plt.title("Polígono de frequência de nicotinina para fumantes")
plt.xlabel("índice das classes de nicotinina")
plt.ylabel("Freq. Relativa")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(range(8),tabelaFreqNaoFumantes['freq. Relativa (%)'],color = 'blue')
plt.title("Poligono de frequência de nicotinina para não fumantes")
plt.xlabel("índice das classes de nicotinina")
plt.grid(True)
plt.show()
```
