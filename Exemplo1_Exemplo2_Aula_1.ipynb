```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Dados
valores = [0,1,5,6,6,0,7,3,5,1,1,1,0,3,5,1,6,4,8,7,2,6,2,1,7,5,8,6,8,3,5,0,4,3,2,5,7,6,5,8]
np.sort(valores)
print(valores)
```

```python
# Tabela de frequÃªncia
frequencias = np.array([valores.count(valor) for valor in range(9)])
frequencias
X = np.arange(9)
frequenciasAcumulada = np.cumsum(frequencias)
frequenciaRelativa = np.round(frequencias/sum(frequencias)*100,1)
df = pd.DataFrame({'freq':frequencias,'freqAcumulada':frequenciasAcumulada,'frequenciaRelativa (%)':frequenciaRelativa},index = X)
df
```

```python
# GrÃ¡fico de barras
plt.figure(figsize = (10,5))
df['freq'].plot(kind = 'barh');
plt.xlabel("FrequÃªncias")
plt.ylabel("valores de X",)
plt.title("DistribuiÃ§Ã£o da freq. Absoluta da variÃ¡vel X")
plt.grid(True)
plt.show()
```

```python
# Dados
Valores = np.array([6.0,10.0,8.2,3.0,3.6,10.0,1.0,1.0,7.0,1.0,6.6,4.0,1.0,3.1,2.3,10.5,6.5,13.0,1.0,7.0,12.0,8.5,10.0,2.6,4.1,
                   5.1,2.3,1.6,3.0,2.0])
len(Valores)
```

```python
# CÃ¡lculo do nÂº total de classes e o cÃ¡lculo da amplitude de cada classe
R = np.max(Valores) - np.min(Valores)
k = np.round(np.sqrt(len(Valores)),0)
h = R/k
print(f"nÂº de classes: {k}")
print(f"amplitude de cada classe: {h}")
```

```python
# CÃ¡lculo das frequÃªncias
print(len(Valores[Valores < 3.4]))
print(len(Valores[Valores < 5.8]) - len(Valores[Valores < 3.4]))
print(len(Valores[Valores < 8.2]) - len(Valores[Valores < 5.8]))
print(len(Valores[Valores < 10.6]) - len(Valores[Valores < 8.2]))
print(len(Valores[Valores <= 13.0]) - len(Valores[Valores < 10.6]))
frequenciasAbsolutas = np.array([13,4,5,6,2])
np.sum(frequenciasAbsolutas)
# determinaÃ§Ã£o das classes de freqâs e frequÃªncias absolutas e relativas
classes = ['[ 1 , 3.4 [','[ 3.4 , 5.8 [', '[ 5.8 , 8.2 [' , '[ 8.2 , 10.6 [','[ 10.6 , 13.0 ]']
frequenciasAcumuladas = np.cumsum(frequenciasAbsolutas)
frequenciasRelativas = np.round(frequenciasAbsolutas/sum(frequenciasAbsolutas)*100,1)
```

```python
#tabela de distribuiÃ§Ã£o de frequÃªncia
df = pd.DataFrame({'freqAbsoluta':frequenciasAbsolutas,'freqAcumulada':frequenciasAcumuladas,'freqRelativas (%)':frequenciasRelativas},index = classes)
df
```
