## Tratamento e análise descritiva dos dados titanic.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('/content/Titanic_dataset.xlsx')
```

```python
# exebindo as 5 primeiras linhas do data-frame
df.head()
```

```python
# verificando o nº de colunas e nº de linhas do data-frame
df.shape
```

```python
# verificando os tipos de dados
df.info()
```

```python
df.columns
```

```python
df.drop(['name','sibsp','parch','ticket','cabin','embarked','boat','body','home.dest'],axis = 1,inplace = True)
```

```python
# consultados as colunas do novo data-frame
df.dtypes
```

```python
df['pclass'] = df['pclass'].astype('category')
df['survived'] = df['survived'].astype('category')
df['sex'] = df['sex'].astype('category')
```

```python
df.dtypes
```

```python
# consultando os dados faltantes
df.isna().sum()
```

```python
# Preenchendo os dados faltantes da coluna age com a média
df['age'].fillna(df['age'].mean(),inplace = True)
# Preenchendo o dados faltante da coluna fare
df['fare'].dropna(inplace = True)
```

```python
# consultando novamento os dados faltantes
df.isna().sum()
```

```python
df['age'] = df['age'].astype('int64')
```

```python
df.dtypes
```

```python
# Separando o data-frame em dois survived = 0 (não sobreviveram) e survived = 1 (sobreviveram)
df_0 = df[df['survived'] == 0]
df_1 = df[df['survived'] == 1]
```

```python
# Comparando as medidas descritivas em relação aos dois grupos
# 0 - não sobreviventes e 1 - sobreviventes
print("Análise descritiva - não sobreviveram\n")
print(df_0.describe())
print('\n')
print("Análise descritiva - sobreviveram\n")
print(df_1.describe())
```

```python
df_0['pclass'].value_counts()
```

```python
df_1['pclass'].value_counts()
```

```python
# gráfico dos sobrevivente e não sobreviventes em relação as classes
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.pie([123,158,528],labels = ['1º classe','2º classe','3º classe'],autopct= '%1.1f%%')
plt.title('Distribuição das classes dos não sobreviventes')
plt.subplot(1,2,2)
plt.pie([200,119,181],labels = ['1º classe','2º classe','3º classe'],autopct = '%1.1f%%')
plt.title('Distribuição das classes dos sobreviventes')
plt.show()
```

```python
print(df_0['sex'].value_counts())
print(df_1['sex'].value_counts())
```

```python
# gráfico dos sobrevivente e não sobreviventes em relação aos sexos
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.pie([682,127],labels = ['homens','mulheres'],autopct= '%1.1f%%')
plt.title('Distribuição das classes dos não sobreviventes')
plt.subplot(1,2,2)
plt.pie([161,339],labels = ['homens','mulheres'],autopct = '%1.1f%%')
plt.title('Distribuição das classes dos sobreviventes')
plt.show()
```

```python
# histogramas em relação as idades dos sobreviventes e não sobreviventes
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.hist(df_0['age'],bins = 5,color ='red')
plt.title('Distribuição das idades dos não sobreviventes.')
plt.xlabel('idades (anos)')
plt.ylabel('frequências absolutas')
plt.subplot(1,2,2)
plt.hist(df_1['age'],bins = 5,color = 'green')
plt.title('Distribuição das idades dos sobreviventes.')
plt.xlabel('idades (anos)')
plt.ylabel('frequências absolutas')
plt.show()
```

```python
# histogramas em relação ao valores das passagens dos sobreviventes e não sobreviventes
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.hist(df_0['fare'],bins = 5,color ='red')
plt.title('Distribuição das passagens dos não sobreviventes.')
plt.xlabel('valor das passagens US$')
plt.ylabel('frequências absolutas')
plt.subplot(1,2,2)
plt.hist(df_1['fare'],bins = 5,color = 'green')
plt.title('Distribuição das idades dos sobreviventes.')
plt.xlabel('valor das passagens US$')
plt.ylabel('frequências absolutas')
plt.show()
```
