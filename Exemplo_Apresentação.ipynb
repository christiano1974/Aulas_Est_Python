```python
# importação dos pacotes padas, numpy e matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# importação dos dados
dados = pd.read_csv('/content/anos_salarios.csv',sep =";",encoding='latin-1')
```

```python
# Verificação de dados faltantes
dados.isnull().sum()
```

```python
# Verificação de dados duplicados
dados.duplicated().sum()
```

```python
# verificação do números de linhas e colunas
dados.shape
```

```python
# visualização dos 5 primeiro registro do data-frame
dados.head()
```

```python
dados.corr()
```

```python
# Gráfico de dispersão
plt.figure(figsize = (10,5))
plt.scatter(dados.iloc[:,0],dados.iloc[:,1])
plt.title('Gráfico de dispersão entre anos de experiência e salário.')
plt.xlabel('Anos de experiência')
plt.ylabel('Salário')
plt.show()
```

## observação
Temos um ponto outline que corresponde a pessoa com 12 anos de experiência indicado pelo gráfico de dispersão.
Este ponto outiline afeta a relação de correlação bem como o ajuste do modelo de regressão linear.

```python
# Fazendo um gráfico boxplot em relação ao salário
plt.figure(figsize = (10,5))
plt.boxplot(dados.iloc[:,1])
plt.title('Boxplot do salário.')
plt.ylabel('Salários')
plt.show()
```

```python
# Determinação da linha do ponto outline
dados[dados['anos de exeperência']==12]
```

```python
# Eliminação do ponto outline
dados.drop(index = 33,axis = 0,inplace = True)
```

```python
# Verificação
print(dados.shape)
dados.tail()
```

```python
# Gráfico de dispersão sem ponto outline
plt.figure(figsize = (10,5))
plt.scatter(dados.iloc[:,0],dados.iloc[:,1])
plt.title('Gráfico de dispersão entre anos de experiência e salário.')
plt.xlabel('Anos de experiência')
plt.ylabel('Salário')
plt.show()
```

```python
# Fazendo um gráfico boxplot em relação ao salário sem pontos outline
plt.figure(figsize = (10,5))
plt.boxplot(dados.iloc[:,1])
plt.title('Boxplot do salário.')
plt.ylabel('Salários')
plt.show()
```
